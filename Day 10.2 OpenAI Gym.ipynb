```python
import gym # Importing OpenAI gym
env = gym.make('MountainCar-v0') # Importing Mountain Car environment
observation = env.reset() # Reset the environment and returns an intial o
for t in range(100):  # Running instance for 100 timesteps
    env.render() # Displays the pop-up window
    print (observation)
    action = env.action_space.sample() # Action_space tells the range o
    obs, reward, terminated, done, info = env.step(action) 
        #a) observation : An object that represents the observation of th
        #b) reward : The amount of reward an agent can achieve
        #c) done : A value that tells whether the environment can be res
    print (obs, reward, terminated, done, info)
    if done:
        print("Finished after {} timesteps".format(t+1))
        break # Exit
```
